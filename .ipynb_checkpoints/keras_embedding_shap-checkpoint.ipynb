# Keras Embeddings and SHAP

We will show how to fit a deep neural net with categorical embeddings in Keras/Tensorflow 2 and how to interprete its results with deepSHAP.

## Load and inspect data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from plotnine.data import diamonds
from plotnine import *
import warnings
warnings.simplefilter('ignore')
```

```python
diamonds.info()
nums = ['price', 'carat']
diamonds[nums].describe()
cats = ['cut', 'color', 'clarity']
# diamonds[cats].describe()
```

### Univariate description of the key variables

```python
print(ggplot(aes(x='value'), data=diamonds[nums].melt()) 
      + geom_histogram(fill='orange') 
      + facet_wrap('~variable', scales='free')
      + ggtitle('Histograms for numeric variables'))

for i, cat in enumerate(cats):
    print(ggplot(aes(x=cat), data=diamonds) 
          + geom_bar(fill='orange')
          + ggtitle(f'Barplot for variable "{cat}"'))
```

### Associations between price and key variables

```python

(ggplot(aes(x='carat', y='price'), data=diamonds) 
 + geom_point(alpha=0.1, shape='.')
 + geom_smooth(color='orange', method='lowess')
 + scale_y_log10()
 + scale_x_log10()
 + ggtitle('log-log-scatterplot of price against carat'))

```
