# Gamma Regression and SHAP

We will use 

- generalized linear model (GLM),
- a boosted tree, and
- a deep neural net

to predict diamonds prices and interprete these models with [SHAP](https://github.com/slundberg/shap). All model techniques will be optimized against Gamma loss, a natural loss function for right skewed data with light tails, like diamonds prices.

## Modeling

### Strategy

The general modeling strategy is the following:

1. Apply common transformations towards an "interpretable" data set. Here, we will stick to a log-transformation of `carat`.
2. Split data into train/test.
3. Wrap model specific transformations in individual preprocessor functions
    - GLM: Dummy coding of `cut`, `clarity`, and `color`
    - Gradient Boosting: Integer coding of above variables.
    - Deep neural net: Integer coding of above variables. Split of data into a dict of numpy arrays. One with scaled dense features (`log_carat`, `color` and `cut`) and one with an embedding variable (`clarity`) for demonstration purpose. Since the latter variable is ordinal as well, we could easily add it as dense feature. In practical applications, one usual encounters unordered categoricals and then, interpretation with SHAP gets tricky. Thus, we want to include this problem here.
4. Fit the models. 
    - GLM: using Gamma regression,
    - Boosted trees: using LightGBM with optimized parameters using 5-fold cross-validation and GridSearch.
    - Deep neural net: using Keras with Tensorflow 2.0 backend and early stopping on internal validation.
5. Compare and interprete the models by [SHAP](https://github.com/slundberg/shap)
    - by looking at deviance gains on the test data set,
    - by studying SHAP importance values, and 
    - by looking at SHAP dependence plots.
   
   Each algorithm uses its "own" SHAP implementation: Kernel Shap for the GLM, TreeSHAP for boosted trees and deepSHAP for the neural net.

### Apply common transformations towards an "interpretable" data set

```python
prep = diamonds[nums + cats].assign(log_carat=np.log(diamonds.carat)).drop(["carat"], axis=1)
prep.head()
```

### Train/test split

```python
from sklearn.model_selection import train_test_split
y = prep["price"]
X = prep.drop("price", axis=1)

y_train, y_test, X_train, X_test = train_test_split(y, X, train_size=0.8, random_state=2098)
```

### GLM

```python
from sklearn.linear_model import TweedieRegressor
```

```python
def prep_glm(df, ohe=None):
    df = df.copy()
    
    cat_vars = df.select_dtypes("category").columns
    df[cat_vars] = df[cat_vars].apply(lambda x: x.cat.codes)
    return df

X_train_gbm = prep_gbm(X_train)
X_test_gbm = prep_gbm(X_test)
```

```python
glm = TweedieRegressor(power=2, alpha=0, link='log')
glm.fit(X_train_glm, y_train)
```

### Gradient boosted trees

Boosted trees with LightGBM! We tune parameters by RandomizedGridSearch with early stopping.

```python
import lightgbm as lgb
```

#### Data interface for LightGBM

```python
def prep_gbm(df):
    df = df.copy()
    cat_vars = df.select_dtypes("category").columns
    df[cat_vars] = df[cat_vars].apply(lambda x: x.cat.codes)
    return df

X_train_gbm = prep_gbm(X_train)
X_test_gbm = prep_gbm(X_test)

dtrain = lgb.Dataset(X_train_gbm, 
                     label=y_train, 
                     free_raw_data=False)
```

#### Optional parameter tuning

```python
tune = True

if tune:
    from sklearn.model_selection import ParameterSampler, ParameterGrid
    
    # Final narrow grid
    grid = ParameterSampler({
        "learning_rate": [0.02],
        "num_leaves": [20, 30, 40],
        'min_child_samples': [10, 15, 20],
        'reg_alpha': [0, 1, 2],
        'reg_lambda': [0, 1],
        'min_split_gain': [0, 0.001],
        'subsample': [0.9, 1],
        'bagging_freq': [1],
        'colsample_bytree': [1],
        'colsample_bynode': [1],
        "verbose": [2],
        "objective": ["gamma"],
        "metric": ["gamma_deviance"]
    }, n_iter=10)
    
    gridsearch_results = pd.DataFrame(columns=["params", "num_boost_round", "perf"])

    print("We are iterating through grid of size", len(grid))
    for g in grid:
        cv_results = lgb.cv(
            g,
            dtrain,
            num_boost_round=10000,
            nfold=5,
            stratified=False,
            early_stopping_rounds=20)
        perf = cv_results['gamma_deviance-mean']
        new_row = {"params": [g], "num_boost_round": len(perf), "perf": perf[-1]}
        print(new_row)
        gridsearch_results = gridsearch_results.append(new_row, ignore_index=True)  
        gridsearch_results.to_json("regression_gamma_shap/gridsearch_diamonds.txt")
```

#### Load grid

```python
gridsearch_results = pd.read_json("regression_gamma_shap/gridsearch_diamonds.txt")
gridsearch_results.sort_values("perf", inplace=True)
gridsearch_results.head()
best = gridsearch_results.iloc[0]
```

#### Fit model

```python
gbm = lgb.train(best.params[0], 
                dtrain, 
                num_boost_round=best.num_boost_round)
```

## Compare and interprete models

### Performance

We will first compare our models by evaluating Gamma deviance on the test data set.

```python
from sklearn.metrics import mean_gamma_deviance
```

### Importance

```python
import shap

# print the JS visualization code to the notebook
shap.initjs()

# Number of SHAP decompositions
N_SHAP = 2000 

# Initialize tree explainer
gbm_explainer = shap.TreeExplainer(gbm)

# Calculate N_SHAP decompositions
gbm_shap_values = gbm_explainer.shap_values(X_train_gbm[:N_SHAP])
```

```python
shap.summary_plot(gbm_shap_values, 
                  X_train_gbm[:N_SHAP], 
                  plot_type="bar", 
                  color="orange")
```

### Dependence plots

```python
sorted_cols = X_train_gbm.columns[np.argsort(-np.abs(gbm_shap_values).mean(0))]

for name in sorted_cols:
    shap.dependence_plot(name, gbm_shap_values,
                         X_train_gbm[:N_SHAP], display_features=X_train[:N_SHAP], 
                         show=False, alpha=0.25, dot_size=4, x_jitter=0.2, 
                         xmin="percentile(1)", xmax="percentile(99)", title=name)
    plt.title(f"{name}", fontdict={'fontsize': 16, 'weight': 'bold'})
    plt.show()
```
